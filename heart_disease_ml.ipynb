```python
y = y.map({'Yes' :1 , 'No' : 0} )
```

```python
y.astype(float)
```

```python
x['Gender'] = x['Gender'].map({
    'Male':1,'Female':0
})
```

```python
x['Gender'] = x['Gender'].astype(float)
```

```python
x[['Smoking',	'Diabetes',	'Family Heart Disease',	'High Blood Pressure']] = x[['Smoking',	'Diabetes',	'Family Heart Disease',	'High Blood Pressure']].replace(
    {'Yes' :'1' , 'No' : '0'}
)
```

```python
x[['Smoking',	'Diabetes',	'Family Heart Disease',	'High Blood Pressure']] = x[['Smoking',	'Diabetes',	'Family Heart Disease',	'High Blood Pressure']].astype(float)
```

```python
x.info()
y.info()
# now the training and testing dataset are completely rid off string datatype 
```

```python
sc = StandardScaler()
```

```python
x =  sc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
# 70-30 ratio training and testing ratio
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'LR' : LogisticRegression(),
    'RFC' :RandomForestClassifier(),
    'SVC' : SVC(),
    'KNNC' : KNeighborsClassifier()
}
#classifer are been used as final outcome is 0/1
```

```python
for name , model in models.items():
    model.fit(x_train,y_train)
    # print(f'tranning accuracy of {name} :{model.score(x_train, y_train)}')
    y_predict_i = model.predict(x_test)
    print(f'accuracy of {name} :  {accuracy_score(y_predict_i ,y_test)*100:.2f} %')
    # print(classification_report(y_predict_i, y_test))
    
# from this knn has less accuracy than others 
#remainig all 3 are having the same accuracy 
```

```python
#optimised lR model using GridSearchCv
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr_model = LogisticRegression()
# - For small datasets, 'liblinear' is a good choice, whereas 'sag'
#       and 'saga' are faster for large ones;
#     - For :term:`multiclass` problems, all solvers except 'liblinear' minimize the
#       full multinomial loss;
#     - 'liblinear' can only handle binary classification by default. To apply a
#       one-versus-rest scheme for the multiclass setting one can wrap it with the
#       :class:`~sklearn.multiclass.OneVsRestClassifier`.
#     - 'newton-cholesky' is a good choice for
#       `n_samples` >> `n_features * n_classes`, especially with one-hot encoded
#       categorical features with rare categories. Be aware that the memory usage
#       of this solver has a quadratic dependency on `n_features * n_classes`
#       because it explicitly computes the full Hessian matrix.

# C â†’ inverse of regularization strength (higher C = weaker regularization).
# solver â†’ optimization method (try multiple to see which converges best).
# cv=5 â†’ 5-fold cross-validation ensures stable performance estimation.
# scoring='accuracy' â†’ you can also use 'roc_auc' if dataset is imbalanced.
```

```python
lr_paramgrid = {
    # hyper parameters
    'C':[0.01 ,0.1 ,1,10,100],                                # Regularization strength
    'solver' :['liblinear' ,'saga','newton-cg' ,'lbfgs'],  # Optimization algorithms
    'penalty' :['l2']                                       # l2 !=12
}
```

```python
lr_GCV_obj =   GridSearchCV(lr_model ,param_grid=lr_paramgrid ,cv=5 ,n_jobs=-1 ,scoring='accuracy')
```

```python
lr_GCV_obj.fit(x_train ,y_train)
```

```python
print(lr_GCV_obj.best_estimator_)
```

```python
print(lr_GCV_obj.best_params_)
```

```python
lr_GCv_tuned_model = lr_GCV_obj.best_params_
```

```python
print(f'{lr_GCV_obj.best_score_*100:.2f} %')
```

```python
# The tuned logistic regression model got 80.17 % where normal logistic regression got only 79.73 %
```
